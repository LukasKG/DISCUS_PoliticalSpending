```python
def get_accounted_output(name,accounted,unaccounted,overaccounted):
    acc_total = sum([accounted,unaccounted,overaccounted])
    return '{:<55s}'.format(name+':') \
            + '{:<40s}'.format('accounted = £'+'{0:.2f}'.format(accounted)+' ('+'{0:.1f}'.format((100*accounted/acc_total) if acc_total>0 else 0)+'%)') \
            + '{:<40s}'.format('| un-accounted = £'+'{0:.2f}'.format(unaccounted)+' ('+'{0:.1f}'.format((100*unaccounted/acc_total) if acc_total>0 else 0)+'%)') \
            + '{:<40s}'.format('| over-accounted = £'+'{0:.2f}'.format(overaccounted)+' ('+'{0:.1f}'.format((100*overaccounted/acc_total) if acc_total>0 else 0)+'%)')


output=''
parties = pd.concat((data_coding['Party'],data_invoice['RegulatedEntityName'])).unique()
#categories = data_coding['Expense Category'].unique()

total_accounted = 0
total_unaccounted = 0
total_overaccounted = 0

positions = []

for party in parties:
    print(f"Party: {party}")
    #output += '\n####################################################################################\n  '+party+'\n####################################################################################\n\n'
    sub_party_coding = data_coding[ data_coding['Party']==party ]
    sub_party_invoice = data_invoice[ data_invoice['RegulatedEntityName']==party ]
    
    party_accounted = 0
    party_unaccounted = 0
    party_overaccounted = 0
    
    suppliers = pd.concat((sub_party_coding['Supplier'], sub_party_invoice['SupplierName'])).unique()
    
    for supplier in suppliers:
        sub_sup_coding = sub_party_coding[ sub_party_coding['Supplier']==supplier ]
        sub_sup_invoice = sub_party_invoice[ sub_party_invoice['SupplierName']==supplier ]
        
        spent = sub_sup_coding['Total Spend'].sum()
        filed = sub_sup_invoice['TotalExpenditure'].sum()
        

        accounted = min(spent,filed)

        if spent-filed > 0.1:
            unaccounted = spent-filed
        else:
            unaccounted = 0

        if filed-spent > 0.1:
            overaccounted = filed-spent
        else:
            overaccounted = 0

        positions.append([party, supplier, accounted, unaccounted, overaccounted])
        #print(get_accounted_output(f"{party} - {supplier} ({sub_sup_coding.shape[0]} C | {sub_sup_invoice.shape[0]} I)", accounted, unaccounted, overaccounted))
        
        party_accounted += accounted
        party_unaccounted += unaccounted
        party_overaccounted += overaccounted

    #output += '-'*170+'\n'
    output += get_accounted_output(party,party_accounted,party_unaccounted,party_overaccounted)+'\n'
    
    total_accounted += party_accounted
    total_unaccounted += party_unaccounted
    total_overaccounted += party_overaccounted



```

```python
print(get_accounted_output('Overall',total_accounted,total_unaccounted,total_overaccounted))
print(output)
```

```python
df = pd.DataFrame(positions, columns=['Party', 'Supplier', 'Accounted', 'Unaccounted', 'Over-Accounted']) 

#df = df.loc[df['Accounted']==0]
df = df.loc[df['Party']=='Reform UK']
```
